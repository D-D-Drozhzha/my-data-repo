```python
import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt

def boot_func(series, stat, reps=5000, seed=2112):
    boot_data = []
    np.random.seed(seed=seed)
    for _ in range(reps):
        sample_data = series.sample(frac=1, replace=True)
        sample_stat = stat(sample_data)
        boot_data.append(sample_stat)
    return np.array(boot_data)

boot_means = boot_func(data_50['normal'], np.mean, reps=5000, seed=2112)

k2, p_value = st.normaltest(boot_means)
print("p-value:", round(p_value, 4))

plt.hist(boot_means, bins=30, edgecolor='black')
plt.title("Bootstrap распределение среднего (normal)")
plt.xlabel("Mean")
plt.ylabel("Frequency")
plt.show()

st.probplot(boot_means, dist="norm", plot=plt)
plt.title("QQ-plot для bootstrap среднего (normal)")
plt.show()
```

```python
def boot_func(series, stat, reps=5000, seed=2112):
    boot_data = []
    np.random.seed(seed=seed)
    for _ in range(reps):
        sample_data = series.sample(frac=1, replace=True)
        sample_stat = stat(sample_data)
        boot_data.append(sample_stat)
    return np.array(boot_data)

def bootstrap_bias(series, stat, reps=5000, seed=2112):
    boot_stats = boot_func(series, stat, reps=reps, seed=seed)
    return boot_stats.mean() - stat(series)


bias_mean = {}
bias_median = {}
for col in data_50.columns:
    bias_mean[col] = bootstrap_bias(data_50[col], np.mean)
    bias_median[col] = bootstrap_bias(data_50[col], np.median)

print("Смещения для среднего:", bias_mean)

print("Смещения для медианы:", bias_median)
```

```python
min_bias_mean = min(bias_mean.values(), key=lambda x: abs(x))
min_bias_median = min(bias_median.values(), key=lambda x: abs(x))

print("Минимальное смещение для среднего:", round(abs(min_bias_mean), 4))
print("Минимальное смещение для медианы:", round(abs(min_bias_median), 4))
```
