```python
import numpy as np
import pandas as pd
from scipy.stats import norm

# Функция бутстрапа
def two_sample_boot(x, y, stat, reps=5000, seed=2112):
    boot_data = []
    np.random.seed(seed=seed)
    for _ in range(reps):
        sample_x = x.sample(frac=1, replace=True)
        sample_y = y.sample(frac=1, replace=True)
        sample_stat_diff = stat(sample_x) - stat(sample_y)
        boot_data.append(sample_stat_diff)
    return np.array(boot_data)

# Функция доверительного интервала
def boot_confidence_interval(x, y, stat, reps=5000, alpha=0.05):
    boots = two_sample_boot(x, y, stat, reps=reps)
    stat_hat = stat(x) - stat(y)  # наблюдаемая разница статистик
    se_hat = boots.std()          # стандартная ошибка через бутстрап
    z = norm.ppf(1 - alpha/2)     # 1.96 для 95%-го ДИ
    ci_low = stat_hat - z * se_hat
    ci_high = stat_hat + z * se_hat
    return stat_hat, (ci_low, ci_high)

# Задаём пары и статистики
pairs = [("control", "test_1"), ("control", "test_2"), ("control", "test_3")]
stats = {"mean": np.mean, "median": np.median, "std": lambda x: np.std(x, ddof=0)}

results = []
for a, b in pairs:
    for stat_name, stat_func in stats.items():
        diff, (low, high) = boot_confidence_interval(data_two_sample[a], data_two_sample[b], stat_func)
        significant = "YES" if not (low <= 0 <= high) else "NO"
        results.append([f"{a} - {b}", stat_name, round(diff, 3), round(low, 3), round(high, 3), significant])

results_df = pd.DataFrame(results, columns=["Сравнение", "Статистика", "Разница", "ДИ низ", "ДИ верх", "Значимо?"])

# Вывод
print(results_df.to_string(index=False))
```

```python
import numpy as np
import pandas as pd

# Функция бутстрапа
def two_sample_boot(x, y, stat, reps=5000, seed=2112):
    boot_data = []
    np.random.seed(seed=seed)
    for _ in range(reps):
        sample_x = x.sample(frac=1, replace=True)
        sample_y = y.sample(frac=1, replace=True)
        sample_stat_diff = stat(sample_x) - stat(sample_y)
        boot_data.append(sample_stat_diff)
    return np.array(boot_data)

# Функция процентильного доверительного интервала
def boot_percentile_interval(x, y, stat, reps=5000, alpha=0.05):
    boots = two_sample_boot(x, y, stat, reps=reps)
    lower = np.percentile(boots, 100 * (alpha/2))
    upper = np.percentile(boots, 100 * (1 - alpha/2))
    return lower, upper

# Загружаем ваши данные (замените путь на свой CSV)
data_two_sample = pd.read_csv("data_two_sample.csv")

# Считаем процентильный интервал для std между control и test_3
ci_low, ci_high = boot_percentile_interval(
    data_two_sample["control"], 
    data_two_sample["test_3"],
    lambda x: np.std(x, ddof=0),   # стандартное отклонение без поправки
    reps=5000, 
    alpha=0.05
)

print("95% процентильный ДИ для разницы std (control - test_3):")
print("Нижняя граница:", round(ci_low, 4))
print("Верхняя граница:", round(ci_high, 4))
```

```python
import numpy as np
import pandas as pd
import scipy.stats as st

def boot_bca(x, y, stat, conf_level=0.95, reps=5000, seed=2112):
    boot_data = []
    np.random.seed(seed=seed)
    sample_stat = stat(x) - stat(y)  # выборочное различие

    for _ in range(reps):
        sample_x = x.sample(frac=1, replace=True)
        sample_y = y.sample(frac=1, replace=True)
        sample_stat_diff = stat(sample_x) - stat(sample_y)
        boot_data.append(sample_stat_diff)

    boot_data = np.array(boot_data)

    # Смещение z0
    z0 = st.norm.ppf(np.mean(boot_data < sample_stat))

    # Jackknife для ускорения
    part_1 = []
    part_2 = []
    for i in range(len(x)):
        part_1.append(stat(np.delete(x.to_numpy(), i)) - stat(y.to_numpy()))
    for i in range(len(y)):
        part_2.append(stat(x.to_numpy()) - stat(np.delete(y.to_numpy(), i)))

    jack_samples = np.concatenate([part_1, part_2])
    uu = np.mean(jack_samples) - jack_samples
    acc = np.sum(uu**3) / (6 * (np.sum(uu**2))**1.5)

    # Итоговый интервал
    alpha = 1 - conf_level
    zalpha = st.norm.ppf([alpha/2, 1 - alpha/2])
    new_quants = st.norm.cdf(z0 + (z0 + zalpha) / (1 - acc * (z0 + zalpha)))
    conf_lo, conf_hi = np.quantile(boot_data, new_quants)

    return round(conf_lo, 4), round(conf_hi, 4), round(z0, 4), round(acc, 4)

# Загружаем данные (замените путь на ваш файл)
data_two_sample = pd.read_csv("data_two_sample.csv")

# Считаем BCa-интервал для std (control - test_3)
ci_low, ci_high, z0, acc = boot_bca(
    data_two_sample["control"],
    data_two_sample["test_3"],
    lambda x: np.std(x, ddof=0),
    conf_level=0.95,
    reps=5000
)

print("95% BCa ДИ для разницы std (control - test_3):")
print("Нижняя граница:", ci_low)
print("Верхняя граница:", ci_high)
print("Смещение z0:", z0)
print("Ускорение acc:", acc)
```

```python
import numpy as np
import pandas as pd
import scipy.stats as st

# --- наши функции ---
def two_sample_boot(x, y, stat, reps=5000, seed=2112):
    boot_data = []
    np.random.seed(seed=seed)
    for _ in range(reps):
        sample_x = x.sample(frac=1, replace=True)
        sample_y = y.sample(frac=1, replace=True)
        sample_stat_diff = stat(sample_x) - stat(sample_y)
        boot_data.append(sample_stat_diff)
    return np.array(boot_data)

def normal_interval(x, y, stat, reps=5000, alpha=0.05):
    boots = two_sample_boot(x, y, stat, reps=reps)
    stat_hat = stat(x) - stat(y)
    se_hat = boots.std()
    z = st.norm.ppf(1 - alpha/2)
    return stat_hat, stat_hat - z*se_hat, stat_hat + z*se_hat

def percentile_interval(x, y, stat, reps=5000, alpha=0.05):
    boots = two_sample_boot(x, y, stat, reps=reps)
    stat_hat = stat(x) - stat(y)
    low = np.percentile(boots, 100*alpha/2)
    high = np.percentile(boots, 100*(1-alpha/2))
    return stat_hat, low, high

def bca_interval(x, y, stat, conf_level=0.95, reps=5000, seed=2112):
    boot_data = []
    np.random.seed(seed=seed)
    stat_hat = stat(x) - stat(y)
    for _ in range(reps):
        sx = x.sample(frac=1, replace=True)
        sy = y.sample(frac=1, replace=True)
        boot_data.append(stat(sx) - stat(sy))
    boot_data = np.array(boot_data)
    z0 = st.norm.ppf(np.mean(boot_data < stat_hat))
    # jackknife
    jack = []
    for i in range(len(x)):
        jack.append(stat(np.delete(x.to_numpy(), i)) - stat(y.to_numpy()))
    for i in range(len(y)):
        jack.append(stat(x.to_numpy()) - stat(np.delete(y.to_numpy(), i)))
    jack = np.array(jack)
    uu = np.mean(jack) - jack
    acc = np.sum(uu**3) / (6*(np.sum(uu**2))**1.5)
    alpha = 1 - conf_level
    zalpha = st.norm.ppf([alpha/2, 1-alpha/2])
    new_quants = st.norm.cdf(z0 + (z0+zalpha)/(1-acc*(z0+zalpha)))
    low, high = np.quantile(boot_data, new_quants)
    return stat_hat, low, high

# --- расчёты ---
pairs = [("control","test_1"),("control","test_2"),("control","test_3")]
stats = {"mean": np.mean, "median": np.median, "std": lambda x: np.std(x, ddof=0)}

methods = {"Normal": normal_interval, "Percentile": percentile_interval, "BCa": bca_interval}

results = []
for a,b in pairs:
    for stat_name, stat_func in stats.items():
        for method_name, method_func in methods.items():
            stat_hat, ci_low, ci_high = method_func(data_two_sample[a], data_two_sample[b], stat_func)
            length = ci_high - ci_low
            asym = (ci_high - stat_hat) / (stat_hat - ci_low) if (stat_hat - ci_low)!=0 else np.nan
            results.append([f"{a}-{b}", stat_name, method_name,
                            round(stat_hat,3), round(ci_low,3), round(ci_high,3),
                            round(length,3), round(asym,3)])

df_results = pd.DataFrame(results, columns=["Сравнение","Статистика","Метод","Оценка","ДИ низ","ДИ верх","Длина","Асимметрия"])
print(df_results.to_string(index=False))
```

```python
import numpy as np
import pandas as pd

def bootstrap_pvalue(x, y, stat, reps=5000, seed=2112):
    boots = two_sample_boot(x, y, stat, reps=reps, seed=seed)
    p_left = np.mean(boots <= 0)
    p_right = np.mean(boots >= 0)
    return 2 * min(p_left, p_right)

pairs = [("control","test_1"),("control","test_2"),("control","test_3")]
stats = {"mean": np.mean, "median": np.median, "std": lambda x: np.std(x, ddof=0)}

results = []
for a,b in pairs:
    for stat_name, stat_func in stats.items():
        pval = bootstrap_pvalue(data_two_sample[a], data_two_sample[b], stat_func, reps=5000)
        results.append([f"{a}-{b}", stat_name, round(pval,4)])

df_pvals = pd.DataFrame(results, columns=["Сравнение","Статистика","p-value"])
print(df_pvals.to_string(index=False))

print("\nМаксимальное p-value:", round(df_pvals["p-value"].max(), 2))
```

```python
import numpy as np
import pandas as pd

def permtest(x, y, stat, reps=5000, seed=2112):
    len_x = len(x)
    true_stat = stat(x) - stat(y)
    full_data = np.concatenate([x.to_numpy(), y.to_numpy()])
    perm_stats = []
    np.random.seed(seed=seed)
    
    for _ in range(reps):
        np.random.shuffle(full_data)
        control = full_data[:len_x]
        test = full_data[len_x:]
        perm_stats.append(stat(control) - stat(test))
    
    pval = np.mean(np.abs(np.array(perm_stats)) >= np.abs(true_stat))
    return np.array(perm_stats), pval

# пары и статистики
pairs = [("control","test_1"),("control","test_2"),("control","test_3")]
stats = {"mean": np.mean, "median": np.median, "std": lambda x: np.std(x, ddof=0)}

results = []
for a,b in pairs:
    for stat_name, stat_func in stats.items():
        _, pval = permtest(data_two_sample[a], data_two_sample[b], stat_func, reps=5000)
        results.append([f"{a}-{b}", stat_name, round(pval,4)])

df_pvals_perm = pd.DataFrame(results, columns=["Сравнение","Статистика","p-value"])
print(df_pvals_perm.to_string(index=False))

print("\nМаксимальное p-value:", round(df_pvals_perm["p-value"].max(), 4))
```
