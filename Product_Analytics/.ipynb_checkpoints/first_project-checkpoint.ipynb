```python
import pandas as pd
import numpy as np
from operator import attrgetter
import matplotlib.pylab as plt
import seaborn as sns
```

```python
## Задание 1.
```

```python
customers = pd.read_csv("olist_customers_dataset.csv")
customers.head()
```

```python
orders = pd.read_csv("olist_orders_dataset.csv", parse_dates=["order_purchase_timestamp", "order_approved_at",
            "order_delivered_carrier_date", "order_delivered_customer_date","order_estimated_delivery_date"])
orders.head()
```

```python
items = pd.read_csv("olist_order_items_dataset.csv")
items.head()
```

```python
orders["order_month"] = orders["order_purchase_timestamp"].dt.to_period("M").dt.to_timestamp()
```

```python
delivered_orders = orders[orders["order_status"] == "delivered"]
```

```python
del_orders_uc = delivered_orders.merge(customers[['customer_id', 'customer_unique_id']], 
                                                         on='customer_id',how='left')
```

```python
del_orders_uc['order_period_day'] = del_orders_uc['order_purchase_timestamp'].dt.to_period('D')
```

```python
del_orders_uc['cohort'] = del_orders_uc.groupby('customer_unique_id')['order_purchase_timestamp'].transform('min').dt.to_period('M')
del_orders_uc['cohort_day'] = del_orders_uc.groupby('customer_unique_id')['order_purchase_timestamp'].transform('min').dt.to_period('D')
```

```python
del_orders_uc['period_number_month'] = np.floor((del_orders_uc.order_period_day - del_orders_uc.cohort_day).apply(attrgetter('n')) / 30)
```

```python
cohorts = del_orders_uc.groupby(['cohort', 'period_number_month']).agg(count_of_customers=('customer_unique_id', 'nunique')).reset_index()
```

```python
cohort_pivot = cohorts.pivot_table(index='cohort', columns='period_number_month', values='count_of_customers')
cohort_pivot
```

```python
plt.figure(figsize=(25, 15))
sns.heatmap(cohort_pivot,annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Cohort analysis")
plt.xlabel("period_number_month")
plt.ylabel("customers_cohort")
plt.show()
```

```python
cohort_size = cohort_pivot[cohort_pivot.columns[0]]
cohort_size
```

```python
retention_matrix = cohort_pivot.divide(cohort_size, axis=0)
retention_matrix
```

```python
plt.figure(figsize=(25, 15))
sns.heatmap(retention_matrix,annot=True, fmt=".2%", cmap="YlGnBu")
plt.title("Cohort analysis")
plt.xlabel("period_number_month")
plt.ylabel("retention_cohort")
plt.show()
```

```python
m1_values = retention_matrix[1].median()
m1_values
```

```python
m3_max_values = retention_matrix[3].idxmax()
max_value = retention_matrix[3].max()
max_value = (max_value*100).round(2)

print(f'{m3_max_values}, значение: {max_value}%')
```

```python
## 3.1. Gross Value
```

```python
delivered_orders_copy = delivered_orders.copy()
```

```python
delivered_orders_copy = delivered_orders_copy.merge(customers[["customer_id","customer_unique_id"]], on="customer_id", how="left")
```

```python
gv = items.merge(delivered_orders_copy[["order_id","order_month","customer_unique_id"]], on="order_id", how="inner")
```

```python
gv_monthly = gv.groupby("order_month")["price"].sum().reset_index(name="GV")
```

```python
gv_monthly
```

```python
## 3.2. Paying customers
```

```python
paying_customers = delivered_orders_copy.groupby("order_month")["customer_unique_id"].nunique().reset_index(name="paying_customers")
```

```python
paying_customers
```

```python
## 3.3. CR_1 (конверсия в первую покупку)
```

```python
## 3.4.Repeat Purchase Rate
```

```python
counts_purchase = delivered_orders_copy.groupby(["order_month", "customer_unique_id"])["order_id"].count()
```

```python
repeat_rate_monthly = ((counts_purchase >= 2).groupby(counts_purchase.index.get_level_values("order_month"))
                       .mean().reset_index(name="RepeatPurchaseRate"))
```

```python
repeat_rate_monthly
```

```python
## 3.5. ARPPU (GV ÷ Paying Customers)
```

```python
ARPPU_monthly = gv_monthly.merge(paying_customers, on="order_month")
```

```python
ARPPU_monthly["ARPPU"] = ARPPU_monthly["GV"] / ARPPU_monthly["paying_customers"]
```

```python
ARPPU_monthly
```

```python
## Vizualization
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14,10))

## GV
axes[0,0].plot(gv_monthly["order_month"], gv_monthly["GV"], marker="o")
axes[0,0].set_title("GV (Сумма заказов)")
axes[0,0].tick_params(axis='x', rotation=45)
axes[0,0].grid(True, linestyle="-", alpha=0.6)

## Paying Customers
axes[0,1].plot(paying_customers["order_month"], paying_customers["paying_customers"], color="orange", marker="o")
axes[0,1].set_title("Paying_customers (Количество платящих клиентов)")
axes[0,1].tick_params(axis='x', rotation=45)
axes[0,1].grid(True, linestyle="-", alpha=0.6)

## Repeat Purchase Rate
axes[1,0].plot(repeat_rate_monthly["order_month"], repeat_rate_monthly["RepeatPurchaseRate"]*100,
               color="green", marker="o")
axes[1,0].set_title("Repeat Purchase Rate (%)")
axes[1,0].tick_params(axis='x', rotation=45)
axes[1,0].grid(True, linestyle="-", alpha=0.6)

## ARPU
axes[1,1].plot(ARPPU_monthly["order_month"], ARPPU_monthly["ARPPU"], color="red", marker="o")
axes[1,1].set_title("ARPPU (Средняя выручка на клиента)")
axes[1,1].tick_params(axis='x', rotation=45)
axes[1,1].grid(True, linestyle="-", alpha=0.6)

plt.tight_layout()
plt.show()
```
